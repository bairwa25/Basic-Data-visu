## Data Visualization Using Python

1. Line Plotting
1. Bar Plotting
1. Histogram Plotting
1. Scatter Plotting

```python
# importing matplotlib module
from matplotlib import pyplot as plt
```

```python
# x-axis values
#   [x1, x2, x3, x4]
x = [5, 2, 9, 4, 7, 8, 2, 6]

# Y-axis values
#   [y1, y2, y3, y4]
y = [10, 25, 8, 4, 2, 6, 3, 9]

plt.xlabel("x - axis")
plt.ylabel("y - axis")

# function for Line Plotting
plt.plot(x, y)
plt.title("Line Plot Visualization")
plt.show()

# fuction for Bar Plotting
plt.bar(x, y)
plt.title("Bar Plot Visualization")
plt.show()

# fuction for Histogram Plotting
plt.hist(y, bins=10)
plt.title("Histogram Plot Visualization")
plt.show()

# fuction for Scatter Plotting
plt.scatter(x,y)
plt.title("Scatter Plot Visualization")
plt.show()
```
